```python
import qrcode

# Generate basic QR code
qr = qrcode.QRCode(
    version=1,
    error_correction=qrcode.constants.ERROR_CORRECT_L,
    box_size=10,
    border=4,
)
qr.add_data("https://www.youtube.com/channel/UCNK47RuwnhT1W7KEq0Pa1mA")
qr.make(fit=True)

img = qr.make_image(fill_color="black", back_color="white")

display(img)
```
